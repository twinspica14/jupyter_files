```python
# dx, dy = np.meshgrid()
#
# _x = np.array([[2,4,5,7,19]])
# _y = np.array([[2,19],[2,19],[2,19],[2,19],[2,19]])
_x = np.linspace(0,10,50)

_y = np.linspace(0,5,20)
# _y = np.arange(2,51,2)
x,y = np.meshgrid(_x,_y) 
# print(_x.shape)
# print(_y.shape)
print("{}\n".format(x))
print("{}".format(y))
# lol = np.dot(_x,_y)
#mesh grid creates matrix such as [xxxx], and [yyyyy] so it's easy to evaulate all points
#example func(x,y) where all value would be evaluated
x.dot(y)
```

```python
# plt.xlim(0,20)
# plt.ylim(0,20)
z = np.sin(x)**10 + np.cos(10+y*x)*np.cos(x)
# plt.contour(x,y,z, colors='black')
# plt.contour(x,y,z, cmap='RdGy')
plt.contourf(x,y,z,20, cmap='RdGy')
plt.colorbar()
```

```python
plt.imshow(z, extent=[0,10,0,5],origin='lower',cmap='RdGy',alpha=0.5)
plt.colorbar()
# plt.axis(aspect='image')
```

```python
plt.figure(figsize=(12,12),dpi=80)
contours = plt.contour(x,y,z,3,color='black')
plt.clabel(contours,inline= True, fontsize=8)
plt.imshow(z,extent=[0,10,0,5],origin='lower',cmap='RdGy',alpha=0.5)
plt.colorbar()
```

# Customizing Plot Legends

# Customize subplot figure by mpl.gridspec.GridSpec(r,c) r and c is size in row and coloumn

```python
x = np.linspace(0,10,100)

r,c = 2,10
plt.figure(figsize=(12,6),dpi=80) #changing dpi we can zoom it, and change frame size using figsize 
mpl.gridspec.GridSpec(r,c)
plt.subplot2grid((r,c),(0,0),colspan=4,rowspan=2)
plt.axis([0,100,-5,5])
plt.plot(np.cos(x),'o',color="yellow",label="cos(x)")
plt.plot(np.sin(x),'-.',color="cyan",label="sine(x)")
# if we want two labels to be side by side we can use ncol=2
plt.legend(loc='best',ncol=2)

plt.subplot2grid((r,c),(0,4),colspan=3,rowspan=1)
plt.plot(np.sin(x))

plt.subplot2grid((r,c),(1,4),colspan=3,rowspan=1)
plt.plot(np.sin(x))

# plt.subplot2grid((r,c),(2,2))
# plt.plot(np.sin(x))

plt.subplot2grid((r,c),(0,7),colspan=3,rowspan=1)
plt.plot(np.sin(x),':',color="red",label="sine(x)")
plt.legend()

plt.subplot2grid((r,c),(1,7),colspan=3,rowspan=1)
plt.plot(np.sin(x),'-.',color="cyan",label="sine(x)")

plt.legend(loc='center',frameon=True)
# plt.subplot2grid((r,c),(2,3))
# plt.plot(np.sin(x))
```
