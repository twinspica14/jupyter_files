# PMF is for disceret and It's probability of occurrence of a random value in a given set of values,PDF is for continuous function in a given range

## Method to find PMF, and It's histogram is clearly shown below

```python
import math

def cohen(obj1,obj2):
    obj = obj1.mean() - obj2.mean()
    var1,var2 = obj1.var(),obj2.var()
    l1,l2 = len(obj1), len(obj2)
    pooled = (l1*var1 + l2*var2)/(l1 + l2)
    d = obj/ math.sqrt(pooled)
    return d
size_effect = cohen(_df4['alcohol'],_df4['fixed acidity'])
print(size_effect)
```

```python
# PMF of distributed series
# Definition of PMF
def find_pdf(object_):
    n = len(object_)
    with open('pdf.csv',mode='w') as lol:
        for f in object_.value_counts():
            pdf = f/n
            print("{} {} ".format(pdf,f),file=lol)
    x =object_.value_counts()
    n = pd.DataFrame({"Values":x.index,"Frequency":x.values})
    m = pd.read_csv("pdf.csv", header=None, sep=" ")
    m.columns = ["PDF", "Frequency","Drop"]
    m = m.drop('Drop',axis=1)
    l = pd.merge(m,n,on="Frequency")
    return l

q4 = find_pdf(_df4['quality'])
q4
```

# Percentile
### It's algorithm how percentile and percentile rank are different
## Percentile rank is from (1,100), where as percentile is actually value of the percentile rank
### Now, we have several method from stats and numpy, and also definition method
## The problem with numpy, Is that it's inefficient, and It doesn't uses series to show value, but instead approximate what in real should be percentile value.

```python

a = np.random.randint(0,100,50)
print(a)

# Method #1 finding percentile value using np.percentile()

s = np.percentile(a,50) #where np.percentile(a, n) where n is percentile, It returns value
print(s)

# MEthod #2 Finding percentile rank which equals to stats.rankdata() and percentileofscore()

def percentile_of_no(series,number):
    """ This Series is number of arrays or list of number """
    count = 0
    for s in series:
        if s <= number:
            count += 1
    p = 100*count/len(series)
    return p

d = percentile_of_no(a,44)

# Method #3

b= (stats.rankdata(a,'average')/len(a))*100

# Method #4

c = stats.percentileofscore(a,44)
print("This is percentily array by .rankdata() {}, This is by percentileofscore {}, This is by definition, rank of the score divided by total number {}".format(b,c,d))

## We can find value using percentile, but here we map it to rank of the value or index

def value_by_percentile(series,percentileRank):
    series.sort()
    index = percentileRank * (len(series)-1) // 100
    return series[index]
k = value_by_percentile(a,c)
print("{} value by using percentile_rank".format(k))
```
